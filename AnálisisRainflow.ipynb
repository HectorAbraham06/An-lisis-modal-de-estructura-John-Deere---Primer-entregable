# Análisis rainflow

```python
import numpy as np
import fatpack
import matplotlib.pyplot as plt
import pandas as pd  
```

## Señal 1

```python

names = ['t', 'y']
```

```python
data =  pd.read_csv('ts1.txt', sep='\t', names=names)
y = np.array(data['y'])
t = np.array(data['t'])

# 2. Extraer picos y valles (Rainflow solo necesita los extremos)
reversals = fatpack.find_reversals(y)
reversals = np.array(reversals)

# 3. Aplicar el algoritmo Rainflow Counting
# Esto devuelve los rangos (amplitudes) de los ciclos encontrados
ranges = fatpack.find_rainflow_ranges(y)

# 4. Análisis Estadístico: Agrupar en un histograma (Bins)
# Definimos los intervalos para clasificar los rangos de aceleración
bins = np.linspace(0, ranges.max(), 10)

# --- Visualización ---
plt.figure(figsize=(6, 4))
plt.hist(ranges, bins=bins)
plt.xlabel('Stress Range')
plt.ylabel('Cycles')
plt.title('Histograma de tensiones para señal 1')

plt.show()

# Señal original


plt.figure(figsize=(6, 4))
plt.plot(t, y, lw=1)
plt.title("Serie Temporal de Aceleración. Señal 1")
plt.xlabel("Tiempo (s)")
plt.ylabel("Aceleración ($m/s^2$)")


plt.show()

intervals = np.linspace(0, max(ranges), 10+1)
stress_ranges = pd.cut(ranges, intervals)

df = stress_ranges.value_counts().reset_index()
df.columns = ['Stress', '# of cycles']
df['Midpoint stress'] = df['Stress'].apply(lambda x: x.mid)
df = df[['Stress', 'Midpoint stress', '# of cycles']]
print(df)
```
